#Implement Gradient Descent Algorithm.

Develop a program in Python to create a Gradient Descent.

```python
import matplotlib.pyplot as plt
```

```python
# Step1: Initaialize Parametrs
x = 2
lr = 0.01
precision = 0.000001
previous_step_size = 1
max_iters = 10000
iters = 0
```

```python
# Step2: Define the gradient of our function
df = lambda x : 2 * (x + 5)
```

```python
# Step3: Run a loop to perform gradient function
gd = []
while previous_step_size > precision and iters < max_iters:
    prev = x
    x = x - lr * df(prev)
    previous_step_size = abs(x - prev)
    iters += 1
    print("Iteration: ", iters, "\tValue: ",x)
    gd.append(x)
```

```python
print("Local Minima: ",x)
```

```python
plt.plot(gd)
```
